# Basics

````python
from skimage import data
import matplotlib.pyplot as plt

# Load an example image from scikit-image
img = data.coffee()

# Show the image
plt.imshow(img)  # pass the image to the matplotlib
plt.axis('off')  # hide axes for a cleaner display
plt.show()       # display the image
````

````python
print(f"Type of the image is: {type(img)}")
print(f"Shape of the image is: {img.shape}")  # (Height, Width, Channels)
````

````python
# Crop the image - array[from:to:step, from:to:step, ...]
img_modified = img.copy()   # make copy to avoid modifying the original img
img_modified = img_modified[:int(img.shape[0] / 2), :int(img.shape[1] / 2)]

# Show the image
plt.imshow(img_modified)  # pass the image to the matplotlib
plt.axis('off')           # hide axes for a cleaner display
plt.show()                # display the image
````

````python
# Make a horizontal stripe - changing the red channel
img_modified[10:20, :, 0] = 255

# Show the image
plt.imshow(img_modified)  # pass the image to the matplotlib
plt.axis('off')           # hide axes for a cleaner display
plt.show()                # display the image
````

````python
# Switch the channels from RGB to BGR
img_modified = img_modified[:,:,::-1]

# Show the image
plt.imshow(img_modified)  # pass the image to the matplotlib
plt.axis('off')           # hide axes for a cleaner display
plt.show()                # display the image
````

````python
# Reduce resolution (4x)
img_modified = img_modified[::4, ::4]

# Show the image
plt.imshow(img_modified)  # pass the image to the matplotlib
plt.axis('off')           # hide axes for a cleaner display
plt.show()                # display the image
````

**Note:**

The **NumPy** package is commonly used for mathematical operations on image, such as manipulating arrays, performing mathematical computations, etc.
However, for more advanced image processing tasks, you should consider using specialized libraries, such as **Pillow** (`PIL` package)

**Let's take a look at the Pillow package:**

````python
from PIL import Image

# Convert the NumPy array to a PIL Image
img_pil = Image.fromarray(img)

# Display the original image (for a single PIL image, you can type the following)
img_pil
````

````python
print(f"Type of the image is: {type(img_pil)}")
print(f"Size of the image is: {img_pil.size}; i.e., width = {img_pil.width}, height = {img_pil.height}")  # [Width, Height]
print(f"Number of channels: {len(img_pil.getbands())}; and channel names: {img_pil.getbands()}")
print(f"I.e., the image is loaded as '{img_pil.mode}'")
````
